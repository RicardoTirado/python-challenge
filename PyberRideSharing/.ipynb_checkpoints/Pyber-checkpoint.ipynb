```python
#Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from statistics import mean, median, mode
```

```python
#load city csv file
city_df = pd.read_csv("generated_data/city_data.csv")
city_df.head()
```

```python
#load ride csv file
ride_df = pd.read_csv("generated_data/ride_data.csv")
ride_df.head()
```

```python
#Merge datasets on "city" so that no data is lost (should be 182 rows)
combined_data = pd.merge(city_df, ride_df, how='outer', on='city')
combined_data.head()
```

```python
#Create variables for pie charts
urban=combined_data[combined_data["type"]="Urban"]
suburban=combined_data[combined_data["type"]="Suburban"]
rural=combined_data[combined_data["type"]="Rural"]
```

```python
#Create avg fares, ride counts, and driver count
urban_fare=urban.groupby(["city"]).mean()["fare"]
urban_ride=urban.groupby(["city"]).count()["ride_id"]
urban_driver=urban.groupby(["city"]).mean()["driver_count"]

suburban_fare=suburban.groupby(["city"]).mean()["fare"]
suburban_ride=suburban.groupby(["city"]).count()["ride_id"]
suburban_driver=suburban.groupby(["city"]).mean()["driver_count"]

rural_fare=rural.groupby(["city"]).mean()["fare"]
rural_ride=rural.groupby(["city"]).count()["ride_id"]
rural_driver=rural.groupby(["city"]).mean()["driver_count"]
```

```python
#Bubble plot chart with variables
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides")
plt.ylabel("Average Fare $")
plt.grid(True)

plt.xlim((0, 38))
plt.scatter(urban_ride,urban_fare,linewidth=1, marker="o", color="gold", alpha=0.7, 
            label="Urban", s=urban_driver*8, edgecolor = "black")
plt.scatter(suburban_ride,suburban_fare,linewidth=1, marker="o", color="lightskyblue", alpha=0.8, 
            label="Suburban",s=suburban_driver*8,edgecolor = "blue" )
plt.scatter(rural_ride,rural_fare,linewidth=1, marker="o", color="lightcoral", alpha=0.9, 
            label="Rural",s=rural_driver*8,edgecolor = "brown")
plt.legend(loc="best")
plt.savefig("Pyber Ride Sharing Data.png")
plt.show()
```

```python
#Total fare by city type
fare_type = combined_data.groupby("type")["fare"].sum().reset_index()
fare_type
```

```python
plt.pie(fare_type["fare"], labels=fare_type["type"],colors=["lightcoral","lightskyblue","gold"],
        shadow=True,explode = (0.1,0.05,.1), autopct= "%1.1f%%")
plt.title('Total Fare by City Type')
plt.savefig("Total Fare by city.png")
plt.show()
```

```python
#Total rides by city type
ride_type = combined_data.groupby("type")["fare"].count().reset_index()
ride_type
```

```python
plt.pie(ride_type["fare"], labels=ride_type["type"],colors=["lightcoral","lightskyblue","gold"],
        shadow=True,explode = (0.1,0.05,.1), autopct= "%1.1f%%")
plt.title('Total Ride by City Type')
plt.savefig("Total Ride by city.png")
plt.show()
```

```python
#Total drivers by city type
driver_type = combined_data.groupby("type")["driver_count"].sum().reset_index()
driver_type
```

```python
plt.pie(driver_type["driver_count"], labels=driver_type["type"],colors=["lightcoral","lightskyblue","gold"],
        shadow=True,explode = (0.1,0.05,.1), autopct= "%1.1f%%")
plt.title('Total Ride by City Type')
plt.savefig("Total Ride by city.png")
plt.show()
```
