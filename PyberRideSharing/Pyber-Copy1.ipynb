```python
#Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from statistics import mean, median, mode
```

```python
#load csv files
city_df = pd.read_csv("generated_data/city_data.csv")
ride_df = pd.read_csv("generated_data/ride_data.csv")
```

```python
#Group ride csv file by 'city' column
ride_group = ride_df.groupby('city')

#Calculate average fare per city  
avg_fare_df = ride_group['fare'].mean()

#Calculate total riders per city
total_riders_df = ride_group['ride_id'].count() 
```

```python
# Merge datasets on "city" so that no data is lost (should be 182 rows)
combined_data_df = pd.merge(avg_fare_df, total_riders_df, how='outer', on='city')
combined_data_df
```

```python
# Merge datasets on "city" so that no data is lost (should be 182 rows)
combined_data_df = pd.merge(city_df, ride_df, how='outer', on='city')
combined_data_df
```

```python
# Create a new dataframe: average fare and total rides per city
AvgFare_TotalRides = combined_data_df.loc[(combined_data_df["fare"].mean()) & (combined_data_df["ride_id"].count())]

# Set the index of this new dataframe by city
AvgFare_TotalRides = AvgFare_TotalRides.set_index("city")

AvgFare_TotalRides.head()
```
